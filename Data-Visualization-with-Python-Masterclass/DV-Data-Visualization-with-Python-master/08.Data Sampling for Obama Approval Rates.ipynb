## plotting approve percent

```python
plt.plot(data_big.year_month, data_big.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```

as per above graph, there are too much data and hard to see the trend.
For this kind of situation, sampling comes in handy.

## Data Sampling

```python
#two ways of getting sample data

sampled1 = data_big.sample(frac=0.1) # fraction of 10% of data
sampled2 = data_big.sample(n=10000) # sample data size 10000
```

```python
plt.plot(sampled1.year_month, sampled1.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```

```python
plt.plot(sampled2.year_month, sampled2.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```
