```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

## find the 10 most populous countries (with most populations)

```python
top_10 = data_2007.sort_values('population', ascending=False).head(10)
top_10
```

## Bar chart using plt.br()

```python
x = range(10)
plt.bar(x, top_10.population / 10 ** 6)

plt.xticks(x, top_10.country, rotation = 'vertical') #plt.xticks(coordinates, actual string values, show this lables vertically)

plt.title('10 Most Populous Countries in 2007')
plt.ylabel('Population in Millions')
plt.show()
```
